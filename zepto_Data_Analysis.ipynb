# **Plotting**

```python
sns.countplot(df["Category"])
```

```python
top_products = df.groupby('name')['quantity'].sum().sort_values(ascending=False).head(10).reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=top_products, x='quantity', y='name', palette='viridis')
plt.title('Top 10 Products by Quantity')
plt.xlabel('Total Quantity')
plt.ylabel('Product Name')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='mrp', y='discountedSellingPrice', hue='Category', alpha=0.7)
plt.title('MRP vs Discounted Selling Price')
plt.xlabel('MRP')
plt.ylabel('Selling Price')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
